```python
import numpy as np
import  matplotlib.pyplot as plt
from linear_regression import LinearRegression
from sklearn.linear_model import LinearRegression as BuiltInLinearRegression
```

```python
np.random.seed(42)
m = 100
x = 2 * np.random.rand(m)
y = 4 + 3 * x + np.random.randn(m)
```

```python
plt.rc('font', size=14)
plt.rc('axes', labelsize=14, titlesize=14)
plt.rc('legend', fontsize=14)
plt.rc('xtick', labelsize=10)
plt.rc('ytick', labelsize=10)
```

```python
plt.plot(x , y , "b.")
plt.xlabel("Feature")
plt.ylabel("Target")
plt.grid()
```

```python
model = LinearRegression()
builtInModel = BuiltInLinearRegression()
```

```python
model.train(x , y)
```

```python
best_fit_line = model.draw_line()
```

```python
plt.plot( x , y , "b.")
plt.plot(best_fit_line , "r" , label= "Predictions")
plt.xlabel("Feature")
plt.ylabel("Target")
plt.grid()
```
